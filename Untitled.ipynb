```python
import os
import numpy as np
import cv2
from pathlib import Path
from glob import glob

import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
import keras

from sklearn.model_selection import train_test_split

from dotenv import load_dotenv
```

```python
load_dotenv()
env_path = Path('.')/'.env'
load_dotenv(dotenv_path=env_path)
path = os.getenv('DATA_PATH')
```

```python
def load_data(path, split=0.1):
    images = sorted(glob(os.path.join(path, "images/*")))
    masks = sorted(glob(os.path.join(path, "l_mask/*")))
    #masks = sorted(glob(os.path.join(path, "s_mask/*")))
    #masks = sorted(glob(os.path.join(path, "mask/*")))

    total_size = len(images)
    print(total_size)
    valid_size = int(split * total_size)
    test_size = int(split * total_size)

    train_x, valid_x = train_test_split(images, test_size=valid_size, random_state=14)
    train_y, valid_y = train_test_split(masks, test_size=valid_size, random_state=14)

    train_x, test_x = train_test_split(train_x, test_size=test_size, random_state=14)
    train_y, test_y = train_test_split(train_y, test_size=test_size, random_state=14)

    return (train_x, train_y), (valid_x, valid_y), (test_x, test_y)
    #return images, masks

#path = 'C:/Users/Berend/ONEDRI~1/Studie/SGGS/data/DataSet'

(train_x, train_y), (valid_x, valid_y), (test_x, test_y) = load_data(path)
#images, masks = load_data(path)
```
